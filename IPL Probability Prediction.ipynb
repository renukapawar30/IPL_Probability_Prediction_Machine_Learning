```python
delivery_df['Score'] = delivery_df[['match_id','Ball_score']].groupby('match_id').cumsum()['Ball_score']
```

```python
delivery_df['target_left'] = (delivery_df['total_runs']+1) - delivery_df['Score']
```

```python
delivery_df['Remaining_ball'] = (120 - ((delivery_df['over']-1)*6 + delivery_df['ball']))
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == '0' else '1').astype('int64')
```

```python
delivery_df
```

```python
delivery_df['Wickets'] = delivery_df[['match_id','player_dismissed']].groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delivery_df['Wickets'] = 10- delivery_df['Wickets']
```

```python
delivery_df
```

- Current Run Rate =

$$Current \ run \ rate =\frac{Runs \ Scores}{Number \ of \ overs \ completed}$$

```python
delivery_df['crr'] = delivery_df['Score']*6/(120 - delivery_df['Remaining_ball'])
```

```python
delivery_df
```

- Required Run Rate =

$$Required \ run \ rate =\frac{Remaining \ Scores \ or \ Traget}{Number \ of \ Overs \ left}$$

```python
delivery_df['rrr'] = delivery_df['target_left']*6/(delivery_df['Remaining_ball'])
```

```python
# creating a column result 1 means - won & 0 means -loss
def result(row):
    if row['batting_team'] == row['winner']:
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

## Splitting the data into Features and Target

```python
X = Model_df.iloc[:,:-1]
y = Model_df.iloc[:,-1]
```

```python
y
```

## Splitting the data to Train and Test

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state =2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])],remainder = 'passthrough')
```

```python
trf
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps= [
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe
```

```python
pipe.fit(X_train,y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
Y_pred
```

```python
print('Accuracy:',accuracy_score(y_test,Y_pred))
```
